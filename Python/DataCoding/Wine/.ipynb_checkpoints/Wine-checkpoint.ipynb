## Relationship between Quality and Features

```python
plt.clf()
nrows, ncols = 4,3
fig, axes = plt.subplots(nrows,ncols,figsize=(16,20))
fig.suptitle("Histograms\n")
for i in range(len(features)):
    row = i//ncols
    col = i - row*ncols
    axes[row][col].set_title(features[i])
    
fig.tight_layout()
plt.show()
```
